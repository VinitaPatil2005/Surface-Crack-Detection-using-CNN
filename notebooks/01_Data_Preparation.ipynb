# Surface Crack Detection using CNN

# Data Preparation

## Objectives

- Load the dataset
- Explore the dataset
- Count images in each class
- Visualize sample images
- Prepare the dataset for CNN training

```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf

print("TensorFlow Version:", tf.__version__)
```

```python
dataset_path = "../data"
```

```python
class_names = sorted(os.listdir(dataset_path))

print("Classes:")
print(class_names)
```

```python
print("Total Classes:", len(class_names))
```

```python
for class_name in class_names:
    
    folder = os.path.join(dataset_path, class_name)
    
    print(f"{class_name}: {len(os.listdir(folder))} images")
```

```python
dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=(120,120),
    batch_size=32
)
```

```python
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=(120,120),
    batch_size=32
)
```

```python
print(dataset.class_names)
```

```python
plt.figure(figsize=(8,8))

for images, labels in dataset.take(1):
    
    for i in range(9):
        
        plt.subplot(3,3,i+1)
        
        plt.imshow(images[i].numpy().astype("uint8"))
        
        plt.title(dataset.class_names[labels[i]])
        
        plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
print("Training Batches :", len(dataset))
print("Validation Batches :", len(validation_dataset))
```
